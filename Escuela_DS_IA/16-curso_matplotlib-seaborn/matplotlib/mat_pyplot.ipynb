```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x, y, 'rx')
plt.plot()
```

```python
plt.plot(x, y, 'rs-')
plt.plot()
```

```python
plt.plot(x, y, 'yD-')
plt.plot()
```

```python
plt.hist(x)
plt.plot()
```

```python
plt.pie(x)
plt.plot()
```

```python
plt.boxplot(x)
plt.show()
```
