```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

Matlab

```python
fig, ax = plt.subplots(figsize = (8,8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'g-')
ax.plot(x, x+3, 'b.-')
ax.plot(x, x+4, 'wo:')
```

Pyplot

```python
fig, ax = plt.subplots(figsize = (8,8))

ax.plot(x, x+1, color='red', linewidth = 5, linestyle = '-')
ax.plot(x, x+2, color='green', linewidth = 8, linestyle = '--', marker='x', markersize=8)
ax.plot(x, x+3, color='#ffCC66', linewidth = 3, linestyle = 'dashed')
ax.plot(x, x+4, color= "#667899", linewidth = 2, linestyle = ':', alpha=0.5)
```
