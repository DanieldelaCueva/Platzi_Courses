```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0,0].plot(x,x**2,'r')
axes[0,1].plot(x,np.sqrt(x),'o')
axes[0,2].plot(x,1/x,'y')
axes[0,3].plot(x,x**3,'g')

axes[1,0].plot(x,np.cos(x),'b')
axes[1,1].plot(x,np.sin(x),'p')
axes[1,2].plot(x,np.tan(x),'g')
axes[1,3].plot(x,np.exp(x),'y')
fig.tight_layout()
```
