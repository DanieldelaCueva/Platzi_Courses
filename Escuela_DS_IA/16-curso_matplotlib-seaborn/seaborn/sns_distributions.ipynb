```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(data=tips, x='tip', cumulative=False, hue='sex', stat='count', multiple='dodge')
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True, bw_adjust=1)
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
```
